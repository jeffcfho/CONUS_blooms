```python
%matplotlib inline
from IPython.display import display

import pandas as pd
import numpy as np
```

```python
# import csvs once
# import phytoplankton count csv
df_phyto = pd.DataFrame.from_csv('../NLA/nla2012_wide_phytoplankton_count_02122014.csv')
# import wide profile csv
df_profiles = pd.DataFrame.from_csv('../NLA/nla2012_wide_profile_08232016.csv')
```

```python
# Calculate cyanobacterial biovoume by site and visit
phyto_count = df_phyto
# change indices and explore data
phyto_count.reset_index(inplace=True)
# display(phyto_count['SITE_ID'].value_counts())
# display(phyto_count['VISIT_NO'].value_counts())
phyto_count.set_index(['SITE_ID','VISIT_NO'],inplace=True)
# display(phyto_count.loc['NLA12_AR-101']) #113 rows
# display(phyto_count.loc['NLA12_MT-104']) #3 rows

# look at values for the different phytoplankton descriptions. value_counts() on 'genus'?
# for col in ['ALGAL_GROUP','PHYLUM','GENUS']:
#     print(col)
#     display(phyto_count.loc[phyto_count['ALGAL_GROUP']=='BLUE-GREEN ALGAE',col].value_counts())
# PHYLUM=='CYANOPHYTA' OR ALGAL_GROUP=='BLUE-GREEN ALGAE'
# once decided column, subset to 'cyanobacteria' (might also want to get different cyano taxa)
# seem to be a lot of 'GENUS' taxa. Look at and see the ones used by Rigosi 2014?

# subsample to necessary rows:
phyto_count = phyto_count.loc[phyto_count['SAMPLE_TYPE']=='PHYX'] #only PHYX
cyano_count = phyto_count.loc[phyto_count['ALGAL_GROUP']=='BLUE-GREEN ALGAE'] #cyanos only

# re-order df and aggregate by site_id and visit_no
cyano_count.reset_index(inplace=True)
cyano_sum = cyano_count.groupby(['SITE_ID','VISIT_NO']).sum()
cyano_count.set_index(['SITE_ID','VISIT_NO'],inplace=True)
phyto_count.reset_index(inplace=True) #get total biovolume as well
phyto_sum = phyto_count.groupby(['SITE_ID','VISIT_NO']).sum()
phyto_count.set_index(['SITE_ID','VISIT_NO'],inplace=True)

cybio_sum = pd.concat([cyano_sum['BIOVOLUME'],np.log(cyano_sum['BIOVOLUME']),
                       phyto_sum['BIOVOLUME']],axis=1)
cybio_sum.columns = ['CyanoBV','logCyanoBV','TotalBV']
cybio_sum['CyanoPro'] = cybio_sum['CyanoBV']/cybio_sum['TotalBV']
cybio_sum.dropna(inplace=True)
display(cybio_sum.head())
cybio_sum.hist(figsize=(8,6)) 

# export csv by site/visit
print(len(cybio_sum))
# cybio_sum.to_csv('CyanoBV.csv')
```

```python
# Calculate temperature profile by site and visit
profiles = df_profiles
# change indices and explore data
profiles.reset_index(inplace=True)
# display(profiles['SITE_ID'].value_counts())
# display(profiles['VISIT_NO'].value_counts())
# display(profiles['DEPTH'].value_counts())
# display(profiles['SCUM'].value_counts()) #not many scum measurements
profiles.set_index(['SITE_ID','VISIT_NO'],inplace=True)
# display(profiles.loc['NLA12_NY-0101']) #72 rows
# display(profiles.loc['NLA12_MS-106']) #1 row
# display(profiles.iloc[1:10,31:])
# display(profiles.columns)

#subset of rows/clumns
profiles = profiles.loc[profiles['SAMPLE_TYPE']=='PROF']
# display(profiles)
profile_vars = ['DEPTH','TEMPERATURE','OXYGEN','PH']
profiles_sub = profiles[profile_vars]
# calc density of water, assuming salinity is neglible (Read, 2011 A1.1)
def temp_to_density(temp):
    temp_plus288 = temp + 288.9414
    temp_plus68 = temp + 68.12963
    temp_minus4 = temp - 3.9863
    temp_frac = temp_plus288/(508929.2*temp_plus68)*((temp_minus4)**2)
    return 1000*(1.0-temp_frac)
profiles_sub.loc[:,'DENSITY'] = temp_to_density(profiles_sub.loc[:,'TEMPERATURE'])
# display(profiles_sub.loc['NLA12_AL-101',1.0])

# group profiles by SITE_ID and VISIT_NO
profiles_sub.reset_index(inplace=True)
profiles_gb = profiles_sub.groupby(['SITE_ID','VISIT_NO'])
profiles_sub.set_index(['SITE_ID','VISIT_NO'],inplace=True)
# print(profiles_gb)

#get surface temperature
surfs = profiles_sub.loc[profiles_sub['DEPTH']==0]
# display(surfs)
surfs.reset_index(inplace=True)
surf_temps = surfs.groupby(['SITE_ID','VISIT_NO']).mean()
surfs.set_index(['SITE_ID','VISIT_NO'],inplace=True)
print(len(surf_temps))
display(surf_temps.head())

# export csv by site/visit
# surf_temps.to_csv('SurfTempOxPH.csv')

#get mean temperature
watcol_means = profiles_gb.apply(np.mean)
print(len(watcol_means))
display(watcol_means)

# export csv by site/visit
# watcol_means.to_csv('MeanTempOxPH.csv')
```

```python
#bottom temperature
def get_profile_bot(testp):
    testp_1d = testp.groupby('DEPTH').agg(np.mean) #get 1 meas per depth by taking mean
    return testp_1d.tail(1) # get bottom depth measurement
profiles_bot = profiles_gb.apply(get_profile_bot)
profiles_bot.reset_index(2,inplace=True)
print(len(profiles_bot))

# checking if removing depth==0 measurements were valid based on index site depth:
nla12lakeinfo = pd.DataFrame.from_csv('../NLA/nla12_keyvariables_data.csv')
nla12lakeinfo.reset_index(inplace=True)
nla12lakeinfo.set_index(['SITE_ID','VISIT_NO'],inplace=True)
profiles_bot = profiles_bot.join(nla12lakeinfo['INDEX_SITE_DEPTH'])
# they are! but enforcing any other restriction 
#  (like DEPTH within 0.5 of INDEX_SITE DEPTH) seems too restrictive given NaNs

profiles_bot = profiles_bot.loc[profiles_bot['DEPTH']!=0]
print(len(profiles_bot))
# display(profiles_bot.sort_values('DEPTH'))

# export csv by site/visit
# profiles_bot.to_csv('BotTempOxPH.csv')
```

```python
#thermocline depth: maximum density gradient in water column
##Brunt Vaisala buoyancy frequency, calculated at thermocline: used to estimate
# steepness of thermocline, a key control on vertical mixing.
# High buoyancy frequency means thermocline is steep, and resistance to mixing is strong.
# (Kraemer 2015)
def calc_thermocline_depth(testp):
#     testp = profiles_sub.loc['NLA12_MS-R03'] #for testing
    testp_1d = testp.groupby('DEPTH').agg(np.mean) #get 1 meas per depth by taking mean
#     display(testp_1d)
    if (len(testp_1d)==0) or (len(testp_1d)==1):
        return [np.nan,np.nan]
    testp_1d.reset_index(inplace=True)
    # testp_1d = pd.DataFrame({'DEPTH':[0,1,2,4],'DENSITY':[1000,1001,1003,1006]}) # toy example as in notebook
#     display(testp_1d)
    testp_1d.loc[:,'NEXT_DEPTH'] = testp_1d.loc[:,'DEPTH'] - testp_1d.loc[:,'DEPTH'].diff(-1)
    testp_1d.loc[:,'DEPTH_DELTA'] = testp_1d.loc[:,'DEPTH'] - 0.5*testp_1d.loc[:,'DEPTH'].diff(-1)
    testp_1d.loc[:,'DENSITY_GRAD'] = testp_1d.loc[:,'DENSITY'].diff(-1)/testp_1d.loc[:,'DEPTH'].diff(-1)
    testp_1d.loc[:,'DEPTH_DELTA_DELTA_prev'] = testp_1d.loc[:,'DEPTH_DELTA'].diff(1)
    testp_1d.loc[:,'DEPTH_DELTA_DELTA_next'] = -testp_1d.loc[:,'DEPTH_DELTA'].diff(-1)
    testp_1d.loc[:,'DENSITY_GRAD_DELTA_prev'] = testp_1d.loc[:,'DENSITY_GRAD'].diff(1) #neg after zt
    testp_1d.loc[:,'DENSITY_GRAD_DELTA_next'] = testp_1d.loc[:,'DENSITY_GRAD'].diff(-1) # pos after zt
    testp_1d.loc[:,'DENSITY_GRAD_GRAD_next'] = testp_1d.loc[:,'DEPTH_DELTA_DELTA_next']/testp_1d.loc[:,'DENSITY_GRAD_DELTA_next']
    testp_1d.loc[:,'DENSITY_GRAD_GRAD_prev'] = testp_1d.loc[:,'DEPTH_DELTA_DELTA_prev']/testp_1d.loc[:,'DENSITY_GRAD_DELTA_prev']
    testp_1d.loc[:,'DENSITY_GRAD_GRAD_sum'] = testp_1d.loc[:,'DENSITY_GRAD_GRAD_next'] + testp_1d.loc[:,'DENSITY_GRAD_GRAD_prev']
#     display(testp_1d)
    # get one row corresponding to highest DENSITY_GRAD
#     testp_1d = testp_1d.loc[(testp_1d.loc[:,'DENSITY_GRAD_GRAD_next']>0) & (testp_1d.loc[:,'DENSITY_GRAD_GRAD_prev']>0),:] 
    testp_1d = testp_1d.loc[testp_1d['DENSITY_GRAD']==testp_1d['DENSITY_GRAD'].max()]
#     display(testp_1d)
    if testp_1d.shape[0]>1: # should only be one row, if not something's gone wrong
        print('Error: testp_1d rows>1, taking first thermocline')
#         display(testp_1d)
        testp_1d = testp_1d.head(1) #first thermocline

    if np.isnan(testp_1d.loc[:,'DEPTH_DELTA_DELTA_prev'].values) or \
       np.isnan(testp_1d.loc[:,'DEPTH_DELTA_DELTA_next'].values):
        z_t = testp_1d.loc[:,'DEPTH_DELTA'] #gradient is at surface or at bottom
    else: 
        z_t = testp_1d.loc[:,'NEXT_DEPTH'] * (testp_1d.loc[:,'DENSITY_GRAD_GRAD_next']/testp_1d.loc[:,'DENSITY_GRAD_GRAD_sum']) + \
              testp_1d.loc[:,'DEPTH'] * (testp_1d.loc[:,'DENSITY_GRAD_GRAD_prev']/testp_1d.loc[:,'DENSITY_GRAD_GRAD_sum'])
    #calc buoyancy frequency, units of s-2
    n2 = 9.81/testp_1d.loc[:,'DENSITY'] * testp_1d.loc[:,'DENSITY_GRAD'] 
#     print(z_t,n2)
    z_t = z_t.values[0]
    n2 = n2.values[0]
    return[z_t,n2]

# apply function over whole list
# display(profiles_gb)
tcs = profiles_gb.apply(calc_thermocline_depth) # make sure function doesn't display stuff!!
# display(tcs)

# # testing applying of function
# i=0
# for name,group in profiles_gb:
#     if (i>=625) and (i<630):
#         print name
# #         display(group)
#         print(calc_thermocline_depth(group))
#     i+=1
```
