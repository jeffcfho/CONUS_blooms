```python
%matplotlib inline
from IPython.display import display,Image
import pydotplus

import pandas as pd
pd.set_option('max_colwidth',100)
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from itertools import combinations as comb
```

```python
# import full data set from DataAggregation script
df = pd.DataFrame.from_csv('fullDataset.csv')
df.reset_index(inplace=True)
df.set_index(['SITE_ID','VISIT_NO'],inplace=True)

# rename some columns to make it easier to remove duplicate variable combos later
df=df.rename(columns = {
        'tp_ugl':'N_tp_ugl',
        'tn_ugl':'N_tn_ugl',
        'tp_log':'N_tp_log',
        'tn_log':'N_tn_log',
    })

print(df.columns.values)
print(len(df))
display(df.head())
```

```python
df_clean = df.copy()
#removing 0s for log
print(len(df_clean))
df_clean = df_clean.loc[df_clean['P_crop_kgha']!=0]; print len(df_clean)
df_clean = df_clean.loc[df_clean['P_manu_kgha']!=0]; print len(df_clean)
df_clean = df_clean.loc[df_clean['P_fert_kgha']!=0]; print len(df_clean)
df_clean = df_clean.loc[df_clean['temp_surf']<40]; print len(df_clean) #erroneous values
df_clean = df_clean.loc[df_clean['temp_bot']<40]; print len(df_clean) #erroneous values
df_clean = df_clean.loc[df_clean['strat_buoy_freq_s2']>-4]; print len(df_clean) #erroneous values
df_clean = df_clean.loc[df_clean['strat_buoy_freq_s2']<0.05]; print len(df_clean) #outliers
df_clean.describe()
```

```python
#different dataframes for different dependent variables
df_mic = df_clean.loc[df_clean['mic_ppb']!=0]; print len(df_mic)
df_cybv = df_clean.loc[df_clean['cybv_um3ml']!=0]; print len(df_cybv)
df_cypro = df_clean.copy(); print len(df_cypro)
df_chla = df_clean.loc[df_clean['chla_ugl']!=0]; print len(df_chla)
```

```python
pred_vars = sorted(list(df.columns))
print(len(pred_vars))
print(pred_vars)
for item in ['chla_ugl','mic_ppb','chla_log','mic_log',
             'cybv_um3ml','cybv_log','totbv_u3ml','cypro',
             'DATE_COL','MONTH_COL',
             'trophic_cat']:
    pred_vars.remove(item)
print(len(pred_vars))
print(pred_vars)
```

```python
# model formulation
model = 'cypro' #can be 'chla','cybv','cypro', or 'mic'
# indepvars = ['tp_log','temp_surf']

# get right model set up
dict_meta = {
    'chla' :[df_chla, 'chla_log'],
    'cybv' :[df_cybv, 'cybv_log'],
    'cypro':[df_cypro,'cypro'],
    'mic'  :[df_mic,  'mic_log']
}
df_model = dict_meta[model][0]
depvar = dict_meta[model][1]
# df_model[[depvar]+indepvars].hist()
print(len(df_model))
```

```python
# loop through all n-var models
largest_n = 5
combs_all = []
for n in range(1,largest_n+1):
    combs_all = combs_all+list(comb(pred_vars,n)) #predvars
    
def check_nodupes(tple1):
    N_sum = sum(['N_' in v for v in tple1])
    prec_sum = sum(['prec_' in v for v in tple1])
    P_sum = sum(['P_' in v for v in tple1])
    temp_sum = sum(['temp_' in v for v in tple1])
    strat_sum = sum(['strat_' in v for v in tple1])
    return not any([N_sum>1,prec_sum>1,P_sum>1,temp_sum>1,strat_sum>1])
#remove combs with duplicate variables
combs_all = [tple for tple in combs_all if check_nodupes(tple)]
print(len(combs_all))

bics = []; r2s = []; ns = []
for indepvars in combs_all: 
    indepvars = list(indepvars)
    #run OLS
    y = df_model[depvar]
    X = df_model[indepvars]
    X = sm.add_constant(X)
    est = sm.OLS(y,X,missing='drop').fit()
    # print(est.summary())
#     print(est.bic)
    bics.append(est.bic)
    r2s.append(est.rsquared)
    ns.append(len(indepvars))

#output models
print model
models = pd.DataFrame(zip(combs_all,r2s,bics,ns),columns=['model','R2','BIC','n']).sort_values('BIC')
models.loc[:,'Exp_mi'] = np.exp(-0.5*models.loc[:,'BIC'])
models.loc[:,'prob_mi'] = models.loc[:,'Exp_mi']/models.loc[:,'Exp_mi'].sum()
display(models.head(10))
```

```python
#determine best variables:
pred_var_probs = []
for varname in pred_vars:
    def check_var(tple1):
        name_sum = sum([varname in v for v in tple1])
        return name_sum>0
    pred_var_probs.append(models.loc[models.model.apply(check_var),'prob_mi'].sum())

df_probs = pd.DataFrame(zip(pred_vars,pred_var_probs),
                        columns=['pred_var','prob']).sort_values('prob',ascending=False)
pd.set_option('display.float_format', lambda x: '%.3f' % x)
display(df_probs)
pd.set_option('display.float_format', None)
```
